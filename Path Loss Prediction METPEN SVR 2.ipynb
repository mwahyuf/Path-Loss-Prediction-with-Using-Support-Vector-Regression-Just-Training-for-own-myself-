## Path Loss Prediction Based On SVR Algorithm

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
# Import dataset 

bus_way = pd.read_excel('BusWay.xlsx')
```

```python
# Membuang kolom yang tidak diperlukan
bus_way = bus_way.drop(['Koridor_No', 'Koridor', 'No.', 'Longitude_of_User', 'Latitude_of_User', 'DateTime', 
                   'Serving RSRP_All Logs', 'Serving RSSI_All Logs', 'PCC PUSCH Path Loss_All Logs'], axis=1)
bus_way
```

### Preprocessing

```python
from sklearn.preprocessing import StandardScaler
features = ['LTE Serving Cell Distance (m)', 'Frequency (MHz)', 'Total_Heigh_of_TX (m)', 'Total_Heigh_of_RX (m)', 
            'Lebar Jalan (m)', 'Tinggi Gedung Sekitar (m)', 'Jarak Antar Gedung (m)']

# fitur x (Input)
x = bus_way.loc[:, features].values

# fitur y (Output)
y = bus_way.loc[:, ['PUCCH Path Loss_All Logs']].values
```

```python
print(x)
```

```python
print(y)
```

```python
# Standardisasi data
scaler = StandardScaler()

# Input
x = scaler.fit_transform(x)
print(x)
```

```python
# Output
y = scaler.fit_transform(y)
y_sc = pd.DataFrame(data = y, columns = ['PUCCH Path Loss_All Logs'])
print(y_sc)
```

### Melakukan PCA untuk reduksi

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
principal_bus_way = pd.DataFrame(data = principalComponents, columns = ['Input'])
```

```python
final_bus_way = pd.concat([principal_bus_way, y_sc], axis=1)
print(final_bus_way)            
```

### Melakukan splitting data

```python
from sklearn.model_selection import train_test_split

x = final_bus_way.loc[:, ['Input']].values

y = final_bus_way.loc[:, ['PUCCH Path Loss_All Logs']].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print("x Train Shape :", x_train.shape)
print("x test Shape :", x_test.shape)
print("y train shape :", y_train.shape)
print("y test shape :", y_test.shape)
```

### Menentukan Nilai MAPE, RMSE, MAE, MaxPE

```python
# Untuk MAPE

def MAPE(y_actual,y_predicted):
    mape = np.mean(np.abs((y_actual - y_predicted)/y_actual))*100
    return mape
```

```python
# Untuk RMSE

def RMSE(y_actual, y_predicted):
    subtract = y_predicted - y_actual
    square = subtract ** 2
    mean = square.mean()
    rmse = np.sqrt(mean)
    return rmse
```

```python
# Untuk MAE

def MAE(y_actual, y_predicted):
    mae = np.mean(abs(y_actual - y_predicted))
    return mae
```

```python
# Untuk MaxPE

def MaxPE(y_actual, y_predicted):
    maxpe = np.max(y_actual - y_predicted)
    return maxpe
```

### Model Support Vector Regressor (SVR) untuk mendapatkan hasil MAPE, RMSE, MAE, dan MaxPE

```python
from sklearn.svm import SVR
```

```python
regressor = SVR(kernel='poly')
svr_model = regressor.fit(x_train, y_train)

svr_pred = svr_model.predict(x_test)
```

```python
# Nilai MAPE
svr_mape = MAPE(y_test, svr_pred)
print("MAPE :", svr_mape, "%")
```

```python
# Nilai RMSE
svr_rmse = RMSE(y_test, svr_pred)
print("RMSE :", svr_rmse, "dB")
```

```python
# Nilai MAE
svr_mae = MAE(y_test, svr_pred)
print("MAE :", svr_mae, "dB")
```

```python
# Nilai MaxPE

svr_maxpe = MaxPE(y_test, svr_pred)
print("MaxPE :", svr_maxpe, "dB")
```
